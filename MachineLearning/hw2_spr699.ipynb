# Problem 3

```python
import pandas
from nltk import word_tokenize
data = pandas.read_csv("sentiment140_sample1.csv", header = None, encoding = "ISO-8859-1")
print(data)
```

## Part 1

```python
def get_counts(data):
    counts = {"pos":{}, "neg":{}}
    
    for index, row in data.iterrows():
        if row[0] is 4:
            tweet = row[5]
            for word in word_tokenize(tweet.lower()):
                if word.isalpha() and word not in counts["pos"]:
                    counts["pos"][word] = 1
                elif word.isalpha():
                    counts["pos"][word] += 1
        elif row[0] is 0:
            tweet = row[5]
            for word in word_tokenize(tweet.lower()):
                if word.isalpha() and word not in counts["neg"]:
                    counts["neg"][word] = 1
                elif word.isalpha():
                    counts["neg"][word] += 1
    return counts
            
counts = get_counts(data)
```

```python
print(counts["pos"]["happy"])
print(counts["neg"]["happy"])
print(len(counts["pos"]))
print(len(counts["neg"]))
```

## Part 2

```python
def get_word_prob(counts, word, polarity):
    if word in counts[polarity]:
        freq = counts[polarity][word]
        if polarity is "pos":
            total = len(counts[polarity]) + len(counts["neg"])
        else:
            total = len(counts[polarity]) + len(counts["pos"])
        return freq / total 
    else:
        return 0
    
```

```python
get_word_prob(counts, "happy", "pos")
```

```python
get_word_prob(counts, "sad", "pos")
```

## Part 3

```python
import math

def log_odds_ratio(count_dict, word, polarity):
    if polarity is "pos":
        num = get_word_prob(count_dict, word, polarity)
        denom = get_word_prob(count_dict, word, "neg")
    elif polarity is "neg":
        num = get_word_prob(count_dict, word, polarity)
        denom = get_word_prob(count_dict, word, "pos")
       
    if denom == 0:
        return 10000
    else:
        value = math.log(num / denom)
        return value
```

```python
log_odds_ratio(counts, "great", "pos")
```

```python
log_odds_ratio(counts, "the", "neg")
```

```python
log_odds_ratio(counts, "thsadfasde", "neg")
```

## Part 4

```python
import bisect

def sort_pos_words(data):
    counts = get_counts(data)
    sorted_words = []
    for word in counts["pos"]:
        if counts["pos"][word] > 10:
            bisect.insort(sorted_words, (word, log_odds_ratio(counts, word, "pos")))
    return sorted_words
    
```
